Matplotlib

useful for making plots and graphs

```python
#importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy  to get data to plot
import numpy as np
```

```python
x =np.linspace(0,10,100) #100 values  between 0,10
y = np.sin(x) #sin values of x
z = np.cos(x) #cos values ofx
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting  the data



```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding titles  x, y axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine velue')
plt.title('sine wave')
plt.show()
```

Bar plot

```python
fig  = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages =['c','c++','java','python','c#']
students =[16,17,19,19,10]
ax.bar(languages,students)
plt.xlabel('languages')
plt.ylabel('number of students')
plt.title('students enrolled')
plt.show()
```

Pie chart

```python
fig  = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages =['c','c++','java','python','c#']
students =[16,17,19,19,10]
ax.pie(students,labels = languages, autopct ='%1.1f%%')
plt.show()
```

scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 =plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='b',label='sine')
ax.scatter(x,z,color='r',label='cosine')
```

3D SCATTER PLOT

```python
fig3 = plt.figure()
ax  = plt.axes(projection='3d')
z  =20 *np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
