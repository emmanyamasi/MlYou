```python
#importing  seaborn

import  seaborn as  sns
import matplotlib.pyplot as plt
import numpy as np
import  pandas as pd


```

```python
#total  bill vs  tip dataset
tips  = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a  theme for plots

sns.set_theme()
```

```python
#visualize  the data
sns.relplot(data =tips,x='total_bill',y='tip',col ='time',hue='smoker',style='smoker',size ='size')
```

```python
#house price dataset
from sklearn.datasets  import fetch_california_housing
carlifornia = fetch_california_housing()

cal = pd.DataFrame(carlifornia.data,columns= carlifornia.feature_names)
cal['price']= carlifornia.target
```

```python
cal.head()
```

DISTRIBUTION PLOT -RANGES

```python
sns.displot(cal['HouseAge'])
```

```python
sns.distplot(cal['HouseAge'])
```

CORRELATION MATRIX-positive and negative

```python
cal.corr()
```

```python
#heatmap we have all colums vertically and horizontally

plt.figure(figsize=(10,10))
sns.heatmap(cal.corr(),cbar=True,square =True,fmt='.1f',annot=True,annot_kws={'size':8},cmap="Blues")
```
